```python
def check_null(df):
    print('\ncheck null values \n')
    print('shape df: ', df.shape, '\n')
    ok_columns = [col for col in df if df[col].isna().sum() == 0]
    print('ok columns: ', ok_columns, '\n')
    for col in df:
        if df[col].isna().sum() >0:
            print(col, ':', df[col].isna().sum())
```

```python
input_df_names = {
    'jet_info':'../output/jet_info.csv',
    'IP3D_tracks_truth':'../output/IP3D_tracks_truth_1_uset0_1_ntrk_10_playwithtime_0_trainITK_1_eta_25_4.csv', 
    'IP3D_tracks_reco':'../output/IP3D_tracks_truth_0_uset0_1_ntrk_10_playwithtime_0_trainITK_1_eta_25_4.csv'
    }
```

```python
for i,name in enumerate(input_df_names):
    dict_rename = {'y_pred_HGTD':'y_pred_HGTD_'+name, 'y_pred_ITK':'y_pred_ITK_'+name}
    df_appo = pd.read_csv(input_df_names[name])
    df_appo.rename(columns=dict_rename, inplace=True)
    if i == 0:
        df  = df_appo
    else:
        df = pd.merge(df,df_appo)
```
